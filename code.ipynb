```python
import numpy as np 
df['Magnesium'] = np.log1p(df['Magnesium'])   # log(1+x) to avoid log(0)
df['Malic_Acid'] = np.log1p(df['Malic_Acid'])
df['Color_Intensity'] = np.log1p(df['Color_Intensity'])
```

```python
outlier_summary = {}
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_summary[col] = len(outliers)
for feature, count in outlier_summary.items():
    print(f"{feature}: {count}")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_df)   
    wcss.append(kmeans.inertia_)

plt.plot(range(2, 11), wcss, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
optimal_k = 3   
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(scaled_df)
df['Cluster'] = clusters
print(df.head())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(scaled_df)
plt.scatter(reduced[:, 0], reduced[:, 1], c=clusters, cmap='viridis', s=50)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('K-Means Clustering (PCA)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(scaled_df, clusters)
print(f"Silhouette Score for k={optimal_k}: {score:.3f}")
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

X = scaled_df  

eps_values = [0.5, 1.0, 1.5, 2.0]
min_samples_values = [3, 5, 10, 20]

results = []

for eps in eps_values:
    for min_samples in min_samples_values:
        db = DBSCAN(eps=eps, min_samples=min_samples).fit(X)
        labels = db.labels_
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise = list(labels).count(-1)
        if n_clusters > 1:
            sil_score = silhouette_score(X, labels)
        else:
            sil_score = -1 

        results.append((eps, min_samples, n_clusters, n_noise, sil_score))
        print(f"eps={eps}, min_samples={min_samples}, clusters={n_clusters}, "
              f"noise={n_noise}, silhouette={sil_score:.3f}")
import pandas as pd
results_df = pd.DataFrame(results, columns=["eps", "min_samples", "clusters", "noise", "silhouette"])
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA


X = scaled_df  

def evaluate_clustering(labels, X):
    """Return evaluation metrics for given cluster labels."""
    if len(set(labels)) <= 1:  
            return {
            "Silhouette": None,
            "Davies-Bouldin": None,
            "Calinski-Harabasz": None
        }
    return {
        "Silhouette": silhouette_score(X, labels),
        "Davies-Bouldin": davies_bouldin_score(X, labels),
        "Calinski-Harabasz": calinski_harabasz_score(X, labels)
    }

results = {}


kmeans = KMeans(n_clusters=3, random_state=42)
labels_kmeans = kmeans.fit_predict(X)
results["KMeans"] = evaluate_clustering(labels_kmeans, X)


hier = AgglomerativeClustering(n_clusters=3)
labels_hier = hier.fit_predict(X)
results["Hierarchical"] = evaluate_clustering(labels_hier, X)


dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)
results["DBSCAN"] = evaluate_clustering(labels_dbscan, X)


df_results = pd.DataFrame(results).T
print(df_results)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import matplotlib.pyplot as plt


kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_df)

hc = AgglomerativeClustering(n_clusters=3, linkage="ward")
hc_labels = hc.fit_predict(scaled_df)


db = DBSCAN(eps=1.5, min_samples=5)
db_labels = db.fit_predict(scaled_df)


def evaluate_clustering(X, labels):

    if len(set(labels)) <= 1 or (len(set(labels)) == 2 and -1 in labels):
        return None, None, None
    sil = silhouette_score(X, labels)
    dbi = davies_bouldin_score(X, labels)
    chi = calinski_harabasz_score(X, labels)
    return sil, dbi, chi

silhouette_kmeans, dbi_kmeans, chi_kmeans = evaluate_clustering(scaled_df, kmeans_labels)
silhouette_hc, dbi_hc, chi_hc = evaluate_clustering(scaled_df, hc_labels)
silhouette_db, dbi_db, chi_db = evaluate_clustering(scaled_df, db_labels)


results = {
    "KMeans": {"silhouette": silhouette_kmeans, "dbi": dbi_kmeans, "chi": chi_kmeans},
    "Hierarchical": {"silhouette": silhouette_hc, "dbi": dbi_hc, "chi": chi_hc},
    "DBSCAN": {"silhouette": silhouette_db, "dbi": dbi_db, "chi": chi_db}
}


algorithms = list(results.keys())
silhouette_scores = [results[alg]["silhouette"] for alg in algorithms]
dbi_scores = [results[alg]["dbi"] for alg in algorithms]
chi_scores = [results[alg]["chi"] for alg in algorithms]

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Silhouette
axs[0].bar(algorithms, silhouette_scores, color="skyblue")
axs[0].set_title("Silhouette Score")
axs[0].set_ylim(-1, 1)
axs[0].axhline(0.5, color="red", linestyle="--", label="Good Threshold")
axs[0].legend()

# Davies–Bouldin
axs[1].bar(algorithms, dbi_scores, color="lightgreen")
axs[1].set_title("Davies–Bouldin Index (lower better)")
axs[1].axhline(1.0, color="red", linestyle="--", label="Good Threshold")
axs[1].legend()

# Calinski–Harabasz
axs[2].bar(algorithms, chi_scores, color="salmon")
axs[2].set_title("Calinski–Harabasz Index (higher better)")

plt.suptitle("Clustering Evaluation Metrics Comparison", fontsize=14)
plt.tight_layout()
plt.show()
```
